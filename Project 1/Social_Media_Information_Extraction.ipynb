**Cleaning the text Data**

```
import nltk
nltk.download("all")
```

```
## Preprocess

import string 
punctuation = string.punctuation

from nltk.corpus import stopwords
stopwords = stopwords.words("english")

def _clean(text):
  # convert all the text in Lower case
  text = text.lower()
  # Remove punchuation
  text = "".join(x for x in text if x not in punctuation )

  # Remove stop words
  words = text.split()
  words = [w for w in words if w not in stopwords]

  text = " ".join(words)
  return text

_clean("This is a sample text!!!")  # This is a stop word so its removed.
```

```
tweet['cleaned'] = tweet['text'].apply(_clean)
```

```
## Maximum Keywords are used in tweet

from collections import Counter
complete_text = " ".join(tweet['cleaned'])
words = complete_text.split()
Counter(words).most_common(50)
```

```
## Most used special character.
raw_text = " ".join(tweet['text'])
mentions = [w for w in raw_text.split() if w.startswith("@")]
Counter(mentions).most_common(50)
```

```
raw_text = " ".join(tweet['text'])
hashtags = [w for w in raw_text.split() if w.startswith("#")]
Counter(hashtags).most_common(50)
```

```
from nltk import ngrams
bigrams = ngrams(complete_text.split(), 2)
Counter(bigrams).most_common(20)
```

```
# Print top organization or top ENtity Name
import nltk 
from nltk import word_tokenize, pos_tag, ne_chunk

for text in tweet['text']:
  entities = ne_chunk(pos_tag(word_tokenize(text)))
  for entity in entities:
    if hasattr(entity, "label"):
      print(entity)
```
