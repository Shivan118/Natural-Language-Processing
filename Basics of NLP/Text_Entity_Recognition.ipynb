```
import nltk
```

```
nltk.download('all')
```

```
text = " Devid will visit IBM corp. iffice in lodon"

tokens = nltk.word_tokenize(text)
tokens
```

```
pos = nltk.pos_tag(tokens)
pos
```

```
chunks = nltk.ne_chunk(pos)
for chunk in chunks:
  print(chunk)
```

```
for chunk in chunks:
  if hasattr(chunk, "label"):
    print(chunk)
```
